```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plot example
x=np.linspace(-1,1,50)
y=2*x+1
y1=x**2+1
plt.plot(x,y)
plt.figure()
plt.plot(x,y1,color='red',linewidth=5,linestyle='--')
plt.xlim((-1,2))
plt.ylim((1,2))
plt.xlabel('I am x')

new_ticks=np.linspace(1,2,5,dtype=np.float)
plt.yticks(new_ticks,['A','B','C','D','E'])
print(new_ticks)
plt.show()
```

```python
# scatter plot
n=1024
xx=np.random.normal(0,1,n)
yy=np.random.normal(0,1,n)
T=np.arctan2(yy,xx)
plt.scatter(xx,yy,s=75,c=T,alpha=0.5) #s: size; c=color; alpha:transparency
plt.xlim((-1.5,1.5))
plt.ylim((-2,2))
```

```python
# bar plot
m=12
xxx=np.arange(m)
yyy=np.random.uniform(0.5,1.0,m)
plt.bar(xxx,yyy)
```
